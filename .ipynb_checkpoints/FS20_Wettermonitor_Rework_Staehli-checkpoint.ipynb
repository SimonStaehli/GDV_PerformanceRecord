### Initial Verification of the Data

At first the datasets going to be checked according to their missing values and outliers.

```python
def missing_values(df):
    """function detects missing values and null values
    Returns: Boolean Value True or False for NA-Values and for Null-Values"""
    value_is_na = False
    value_is_null = False
    
    na_values = df.isna().sum()
    null_values = df.isnull().sum()
    
    for column in range(0, len(df.columns)):
        if na_values[column] != 0:
            value_is_na = True
            
        if null_values[column] != 0:
            value_is_null = True
    
    return value_is_na, value_is_null
       
    
def detect_outliers(df, column_name, printer):
    """Detects Outliers (higher than third quantile + IQR and lower than first quantile -IQR) and creates a dataframe with all outliers
    if the boolean variable printer is set true, it prints all values one by one.
    Returns: upper and lower outliers"""
        
    count = df[column_name].shape
    
    median = np.median(df[column_name])
    upper_quantile = np.percentile(df[column_name], 75)
    lower_quantile = np.percentile(df[column_name], 25)
    IQR = upper_quantile - lower_quantile
    
    upper_outliers = df[df[column_name] >= (upper_quantile + (1.5*IQR))][column_name]
    upper_outliers_count = upper_outliers.count()
    lower_outliers = df[df[column_name] <= (lower_quantile - (1.5*IQR))][column_name]
    lower_outliers_count = lower_outliers.count()
    
    if printer == True:
        print("..............................................")
        print("Column: {}".format(column_name).upper())
        print("Upper Outliers = {} / Lower Outliers = {}".format(upper_outliers_count, lower_outliers_count))
    
    return upper_outliers, lower_outliers
        
```

### Outliers in Dataframe and Comparisson of the two Measurement Stations

```python
def plot_outliers(df1, df2):
    """Plots all columns of the dataset"""
    number_of_columns = len(df1.columns)
    iteration = 1
    fig, ax = plt.subplots(number_of_columns, 2, figsize=(15,40))
    set_shift = False
    
    for i in range(len(df1.columns)):
        if df1.columns[i] == 'precipitation':
            set_shift = True
            continue
    
        if set_shift == True:
            p1 = sns.boxplot(df1.iloc[:,i], ax= ax[i-1,0], color='blue', saturation=0.5, )
            p1.set_title('{} - Mythenquai: {}'.format(iteration, df1.columns[i]))
            p1.set_xlabel('')
            iteration += 1
        
            continue
                        
        p1 = sns.boxplot(df1.iloc[:,i], ax= ax[i,0], color='blue', saturation=0.5, )
        p1.set_title('{} - Mythenquai: {}'.format(iteration, df1.columns[i]))
        p1.set_xlabel('')
        
        iteration += 1
    
    iteration = 1
    
    for n in range(len(df2.columns)):
        p2 = sns.boxplot(df2.iloc[:,n], ax=ax[n,1], color='red', saturation=0.5)
        p2.set_title('{} - Tiefenbrunnen: {}'.format(iteration, df2.columns[n]))
        p2.set_xlabel('')
            
        iteration += 1
    
    plt.show()
```

## Seasonalities

To check our dataframe to seasonalities I take the autocorrelation below.


```python
def autocorrelation_df(df, frequency):
    """ This function plots all columns with the Autocorrelation plot function of Pandas. Used correlation coefficient
    for the calculation is Pearson. In order to decrease the size of the dataframe to improve the runtime the frequency of samples can be adjusted.
    Documentation URL:
    Inputs: df, frequency(= frequency in hours of measured sample)
    Returns: Autocorrelation plots for each attribute
    """
    samples_per_hour = 6 #1 measurement every 10 minutes, so 1h consists of 6 measurements
    
    for column in df.columns:
        plot = pd.plotting.autocorrelation_plot(df[column].iloc[::frequency*samples_per_hour]) 
        plt.title("Autocorrelation of {}".format(column), loc = "left")
        plt.xlabel("Time Lag (1 Lag is {}h)".format(frequency))
        plt.show()
```

```python
autocorrelation_df(df_07_18_mq, 12)    
```

The plots above show the autocorrelation of the tested attributes, which means the correlation between itself with a time lag. 
Correlation regarding to time and seasonalities are visible for the attributes, which strongly correlate with the air temperature such as water temperature, windchill, water level, humidity and dew point. For the second group, which correlates with wind, we have zero to almost zero correlation regarding to the time.

#### Hypothesis: Strong winds come from same direction


To check this hypothesis I need to know what wind speeds are denominated as strong winds. Therefore I had a look on the Beaufort-Skala. (https://de.wikipedia.org/wiki/Beaufortskala)
A wind velocity of over 40.74 km/h is rated as strong wind.

```python
df_strong_winds = df_07_18_mq[df_07_18_mq["wind_gust_max_10min"] > 40.74]
df_strong_winds_ = df_07_18_tb[df_07_18_tb["wind_gust_max_10min"] > 40.74]
print("Length of filtered Dataset: {} / {}".format(len(df_strong_winds), len(df_strong_winds_)))

fig, ax = plt.subplots(1,2, figsize=(16, 4))

plot = sns.kdeplot(df_strong_winds["wind_gust_max_10min"], df_strong_winds["wind_direction"], ax=ax[0], cmap='Reds', shade=True)
plot.set_title('Measurements of Wind Direction and Wind Gusts at Station Mythenquai')
plot.set_xlabel('Wind Gusts in km/h')
plot.set_ylabel('Wind Direction in °')

plot = sns.kdeplot(df_strong_winds_["wind_gust_max_10min"], df_strong_winds_["wind_direction"], ax=ax[1], cmap='Blues', shade=True)
plot.set_title('Measurements of Wind Direction and Wind Gusts at Station Tiefenbrunnen')
plot.set_xlabel('Wind Gusts in km/h')
plot.set_ylabel('Wind Direction in °')
plt.show()
```

The most "strong" wind gust were measured around 40 to 50 km/h per hour, whereas the most measured direction lies between 220° and 300°. 
At this point we can see that the most strongest measured winds usually come from the same direction. I think this depends on the valley Zurich is located in. Eventually my hypothesis is true.

#### Hypothesis: Extrem Temperature Values increased over the years

To prove this hypothesis I plot all the maximum Temperature values in a lineplot.

```python
df_extreme = df_07_18_mq["air_temperature"].resample("Y").quantile(0.9).loc["2007":"2018"]
df_extreme_min = df_07_18_mq["air_temperature"].resample("Y").quantile(0.1).loc["2007":"2018"]
df_extreme = np.array(df_extreme.values)
df_extreme_min = np.array(df_extreme_min.values)
years = np.array([i for i in range(2007,len(df_extreme)+2007)])
df_extreme = pd.DataFrame(data = {"year": [*years],
                                  "air_temperature_max": [*df_extreme],
                                 "air_temperature_min": [*df_extreme_min]})

fig, ax = plt.subplots(1,2, figsize= (16,4))
plot = sns.regplot(df_extreme["year"], df_extreme["air_temperature_max"], ax=ax[0], color = "red", ci=95)
plot.set_title("Temperature at 90th Percentile per Year", loc = "left")
plot.set_xlabel("Time in Years")
plot.set_ylabel("Air Temperature in C°")
plot2 = sns.regplot(df_extreme["year"], df_extreme["air_temperature_min"], ax=ax[1], color = "blue", ci=95)
plot2.set_title("Temperature at 10th Percentile per Year", loc = "left")
plot2.set_xlabel("Time in Years")
plot2.set_ylabel("Air Temperature in C°")
plt.show()
```

The plot shows two resampled dataframes with the temperatures at the 90th percentile and the 10th percentile should represent a robuster comparisson than the maximum and minimun temperatures per year as parameter. To identify a maximum  temperature increase or minimum temperature decrease the past years I added a regression line. Additionally there's a confidence intervall of 95% given which shows us the overlap with the truth value estimated. The calculation of the intervall includes the slope of the line, z-score and standard error. 

According to this plot it's visible that the temperature values increased over the past years and the minimum temperature slightl increased as well the last years. Though this plot is to be taken with care because we can see a wide scattering mass. For a proper statement to extreme temperatures I'd need more data to investigate from past years. In this particular case we can say that there's an increasing trend for these yearly temperatures. I declare the hypothesis for these datapoints as true for the temperatuer inrease.
